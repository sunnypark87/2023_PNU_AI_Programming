# Session 1 - type, variable

### Values and Types

```python
a = 5
print(type(a)) # int
b = 3.14
print(type(b)) # float
c = "Hello"
print(type(c)) # str
d = True
print(type(d)) # bool

```

### Numbers

```python
a = 7
b = 100,000,000     # 100_000_000
c = 3.14
d = 0b1101  # binary
e = 0o755   # octal
f = 0x1a      # hexadecimal
print(a, b, c, d, e, f)
```

### Variables

```python
name = "Alice"
age = 30
height = 1.75
is_student = False

print(name, age, height, is_student)
```

### Naming Rules

```python
2things = 1          # invalid
my-name = “alice”    # invalid
Name = ”bob”         # case-sensitive
name = ”charlie”
int = 5              # keyword
float = 3.14         # keyword
```

### Arithmetic Operations

```python
addition = 7 + 3
subtraction = 5.5 - 2.5
multiplication = 3 * 4
division = 10 / 3
floor_division = 10 // 3
modulus = 10 % 3
exponentiation = 2 ** 3
result = (2 + 3) * 4
print(addition, subtraction, multiplication, division, floor_division, modulus, exponentiation, result)
```

### Type Conversion

```python
num_int = 5
num_float = 2.0
result = num_int + num_float                    # “5” + 2.0 TypeError
print("Result data type:", type(result))

num_str = "42"
num_int = int(num_str)
result = num_int + 10
print("Result:", result)
```

### Comparison and Logical Operators

```python
x = 5
y = 10
z = 15
result = x < y and y < z

print(result)
```

### Variables and Assignment

```python
x = 5
y = x + 3
x = 7

print(x, y)
```

### Console Output


```python
print("Hello, World!")
print("a", "b", "c", sep="-", end=" ")
print("d", "e", "f", sep="-")
print("A", "B", "C", sep=" -> ", end=" -> END")
```

# Session 2 - string

### string

```python
single_quote = 'Hello, World!'
double_quote = "Hello, World!"
multiline_string = '''This is a
                      multiline string'''

my_string[0] = "h"     # TypeError

print(single_quote, double_quote, multiline_string)
```

### Concatenation and Repetition


```python
greeting = "Hello" + " " + "World"
line = "-" * 20

print(greeting)
print(line)
```

### String Functions and Methods

```python
s = "hello, world!"
length = len(s)
uppercase = s.upper()
replaced = s.replace("world", "there")
clean_str = s.strip().lower().replace(',', '').replace('!', '')  # chaining

print(length, uppercase, replaced, clean_str)
```

### Indices and Slices


```python
text = "Python"
first_char = text[0]
last_char = text[-1]
slice1 = text[1:4]
slice2 = text[::-1]

print(first_char, last_char, slice1, slice2)
```

### Format Method


```python
name = "Alice"
age = 30
output = "{} is {} years old.".format(name, age)

print(output)
```

### Format Specification and Field Width


```python
x = 10
y = 5
output = "{0:<5} + {1:>5} = {2:^5}".format(x, y, x + y)

print(output)
```

### Floating-Point Numbers and Percentages


```python
pi = 3.14159265
percentage = 0.85
formatted_pi = "{:.2f}".format(pi)
formatted_percentage = "{:.1%}".format(percentage)

print(formatted_pi)
print(formatted_percentage)
```

### Escape Sequences


```python
text = "She said, \"Hello!\""
multiline = "Line 1\nLine 2"

print(text)
print(multiline)
```

### Keyboard Input


```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))

print("Hello,", name, "you are", age, "years old.")
```

### Indentation

```python
def print_hello():
    print("Hello, world!")
    print("This is a properly indented block of code.")

print_hello()
```

### Line Joining


```python
# Explicit line joining
long_string = "This is a very long string that we want to \
break into multiple lines for better readability."

# Implicit line joining
colors = [    "red",    "blue",    "green",]

print(long_string)
print(colors)
```
